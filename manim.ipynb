```
%%manim AddingText --base64
from manimlib.imports import *

class AddingText(Scene):
    #Adding text on the screen
    def construct(self):
        my_first_text=TextMobject("Writing with manim is fun")
        second_line=TextMobject("and easy to do!")
        second_line.next_to(my_first_text,DOWN)
        third_line=TextMobject("for me and you!")
        third_line.next_to(my_first_text,DOWN)

        self.add(my_first_text, second_line)
        self.wait(2)
        self.play(Transform(second_line,third_line))
        self.wait(2)
        second_line.shift(3*DOWN)
        self.play(ApplyMethod(my_first_text.shift,3*UP))
```

```
%%manim Pythagoras --base64
from manimlib.imports import *

class Pythagoras(Scene):
    #A few simple shapes
    def construct(self):

        presquare = Polygon(np.array([-2,2,0]),np.array([2,2,0]),np.array([2,-2,0]),np.array([-2,-2,0]))
        triangle1 = Polygon(np.array([-2,2,0]),np.array([-2,-1,0]),np.array([-1,2,0]),fill_opacity=0.5,stroke_width=1)
        triangle2 = Polygon(np.array([-2,-1,0]),np.array([-2,-2,0]),np.array([1,-2,0]),fill_opacity=0.5,stroke_width=1)
        triangle3 = Polygon(np.array([1,-2,0]),np.array([2,-2,0]),np.array([2,1,0]),fill_opacity=0.5,stroke_width=1)
        triangle4 = Polygon(np.array([2,1,0]),np.array([2,2,0]),np.array([-1,2,0]),fill_opacity=0.5,stroke_width=1)
        presquaregroup = VGroup(presquare,triangle1,triangle2,triangle3,triangle4)
        presquaregroup.generate_target()
        presquaregroup.target.shift(4*LEFT)
        postsquare = presquare.copy()
        postsquare = postsquare.shift(4*RIGHT)
        triangle1.generate_target()
        triangle1.target.shift(4*RIGHT)
        triangle3.generate_target()
        triangle3.target.shift(RIGHT+UP)
        triangle2.generate_target()
        triangle2.target.shift(5*RIGHT)
        triangle4.generate_target()
        triangle4.target.shift(4*RIGHT+3*DOWN)
        asquare = Polygon(np.array([2,-2,0]),np.array([2,-1,0]),np.array([3,-1,0]),np.array([3,-2,0]),color=YELLOW,fill_opacity=0.5)
        bsquare = Polygon(np.array([3,-1,0]),np.array([3,2,0]),np.array([6,2,0]),np.array([6,-1,0]),color=YELLOW,fill_opacity=0.5)
        csquare = Polygon(np.array([-6,-1,0]),np.array([-3,-2,0]),np.array([-2,1,0]),np.array([-5,2,0]),color=YELLOW,fill_opacity=0.5)
        atex = TexMobject(r"a^2")
        atex.move_to(asquare.get_center())
        btex = TexMobject(r"b^2")
        btex.move_to(bsquare.get_center())
        ctex = TexMobject(r"c^2")
        ctex.move_to(csquare.get_center())
        equation = TexMobject("a^2","+","b^2","=","c^2")
        #                       0    1    2    3    4
        equation.to_edge(DOWN)
        atex.generate_target()
        atex.target.move_to(equation[0].get_center())
        btex.generate_target()
        btex.target.move_to(equation[2].get_center())
        ctex.generate_target()
        ctex.target.move_to(equation[4].get_center())
        self.play(ShowCreation(presquare),ShowCreation(triangle1),ShowCreation(triangle2),ShowCreation(triangle3),ShowCreation(triangle4))
        self.wait()
        self.play(MoveToTarget(presquaregroup),ShowCreation(postsquare))
        self.add(postsquare)
        self.play(MoveToTarget(triangle1.copy()),MoveToTarget(triangle3.copy()),MoveToTarget(triangle2.copy()),MoveToTarget(triangle4.copy()))
        self.play(ShowCreation(asquare),ShowCreation(bsquare),ShowCreation(csquare))
        self.play(Write(atex),Write(btex),Write(ctex))
        self.play(MoveToTarget(atex.copy()),MoveToTarget(btex.copy()),MoveToTarget(ctex.copy()))
        self.play(Write(equation))
        self.wait()
      
```

```
%%manim BouncingBall --base64
from manimlib.imports import *

class Box(Rectangle):
    CONFIG = {
        "height": 6,
        "width": FRAME_WIDTH - 2,
        "color": GREEN_C
    }

    def __init__(self, ** kwargs):
        Rectangle.__init__(self, ** kwargs)  # Edges
        self.top = 0.5 * self.height
        self.bottom = -0.5 * self.height
        self.right_edge = 0.5 * self.width
        self.left_edge = -0.5 * self.width
        
class Ball(Circle):
    CONFIG = {
        "radius": 0.4,
        "fill_color": BLUE,
        "fill_opacity": 1,
        "color": BLUE
    }

    def __init__(self, ** kwargs):
        Circle.__init__(self, ** kwargs)
        self.velocity = np.array((2, 0, 0))

    def get_top(self):
        return self.get_center()[1] + self.radius

    def get_bottom(self):
        return self.get_center()[1] - self.radius

    def get_right_edge(self):
        return self.get_center()[0] + self.radius

    def get_left_edge(self):
        return self.get_center()[0] - self.radius

class BouncingBall(Scene):
    CONFIG = {
        "bouncing_time": 5,
        "title_config":{
            "extra_str": "\\it\\  - by EulerTour",
            "buff": 0.2
        }
    }
    def construct(self):
        box = Box()
        ball = Ball()
        self.play(FadeIn(box))
        self.play(FadeIn(ball))

        def update_ball(ball,dt):
            ball.acceleration = np.array((0, -5, 0))
            ball.velocity = ball.velocity + ball.acceleration * dt
            ball.shift(ball.velocity * dt)  # Bounce off ground and roof
            if ball.get_bottom() <= box.bottom*0.96 or \
                    ball.get_top() >= box.top*0.96:
                ball.velocity[1] = -ball.velocity[1]
            # Bounce off walls
            if ball.get_left_edge() <= box.left_edge or \
                    ball.get_right_edge() >= box.right_edge:
                ball.velocity[0] = -ball.velocity[0]

        ball.add_updater(update_ball)
        self.add(ball)

        self.wait(self.bouncing_time)
```

```
%%manim test --base64 -l
from manimlib.imports import *

radius = 3
class Ball(Circle):
  CONFIG = {
      "radius": 0.1,
      "fill_opacity": 1,
      "time": 0,
      "color": YELLOW
  }

class Route(Circle):
  CONFIG = {
      "radius": radius,
      "color": BLUE,
      "stroke_width": 6,
      "num_dashes": 15,
      "positive_space_ratio": 0.5,
      "anchors_span_full_range": False
  }

class test(Scene):
  def construct(self):
    scale = 0.5
    path = DashedVMobject(Route(),num_dashes = 51)
    ball = Ball()
    ball.move_to(np.array([radius,0,0]))
    vvec = Arrow(color=GREEN)
    avec = Arrow(color=RED)
    veltex = TexMobject(r"\vec{v}",color=GREEN)
    acctex = TexMobject(r"\vec{a}",color=RED)
    line = Line(color=PURPLE)
    point = Circle(color=RED,radius=0.05,fill_opacity=1)
    test = Arc
    plane = ComplexPlane()
    plane.add_coordinates()

    def update_ball(ball,dt):
      ball.time = ball.time + dt
      ball.omega = TAU/5
      if ball.time < 1:
        ball.omegatime = 0.5*ball.omega*ball.time**2
      else:
        ball.omegatime = ball.time*ball.omega-0.5*ball.omega
      ball.sin = np.sin(ball.omegatime)
      ball.cos = np.cos(ball.omegatime)
      ball.move_to(ORIGIN + np.array([radius*ball.cos,radius*ball.sin,0]))
    
    def update_vvec (vvec,dt):
      vvec.put_start_and_end_on(
          ball.get_center(),ball.get_center()+scale*np.array
          ([-ball.omega*radius*ball.sin,ball.omega*radius*ball.cos,0]))
    
    def update_avec (avec,dt):
      avec.put_start_and_end_on(
          ball.get_center(),ball.get_center()+scale*np.array(
              [-ball.omega**2*radius*ball.cos,-ball.omega**2*radius*ball.sin,0]
          )
      )

    def update_veltex (veltex,dt):
      veltex.move_to(vvec.get_center_of_mass()+np.array([ball.cos,ball.sin,0])*0.5)

    def update_acctex (acctex,dt):
      acctex.move_to(avec.get_center_of_mass()+np.array([-ball.sin,ball.cos,0])*0.5)

    def update_line (line,dt):
      line.put_start_and_end_on(veltex.get_center(),vvec.get_center())

    def update_point (point,dt):
      point.move_to(vvec.get_center_of_mass())

    self.add(path)
    self.play(ShowCreation(ball)) 
    ball.add_updater(update_ball)
    vvec.add_updater(update_vvec)
    avec.add_updater(update_avec)
    veltex.add_updater(update_veltex)
    acctex.add_updater(update_acctex)
    line.add_updater(update_line)
    point.add_updater(update_point)
    self.add(ball)
    self.wait(2)
    self.play(FadeIn(vvec),FadeIn(avec),FadeIn(veltex),FadeIn(acctex))
    self.play(ShowCreation(plane))
    self.wait(10)
```
